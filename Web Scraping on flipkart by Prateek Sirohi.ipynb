Extracting the details for tshirts having more than 80% discount in clothing from flipkart using web scraping.

```python
url= 'https://www.flipkart.com/clothing-and-accessories/topwear/tshirt/men-tshirt/pr?sid=clo,ash,ank,edy&otracker=categorytree&otracker=nmenu_sub_Men_0_T-Shirts'
result = requests.get(url)

Brand_name = []
Description =[]
MRP =[]
Price =[]
Discount=[]
Link=[]
result.text
soup = BeautifulSoup(result.text, 'html')
box = soup.find('div',class_ ="DOjaWF YJG4Cf")

for i in range(2,10):
    url ='https://www.flipkart.com/clothing-and-accessories/topwear/tshirt/men-tshirt/pr?sid=clo%2Cash%2Cank%2Cedy&otracker=categorytree&otracker=nmenu_sub_Men_0_T-Shirts&page='+ str(i)
    result = requests.get(url)
    soup = BeautifulSoup(result.text,'html')
    next_page = box.find("a", class_ ='_9QVEpD').get('href')
    complete_link = 'https://www.flipkart.com' + next_page
    print(complete_link)
    
    try:
        brand = box.find_all('div', class_='syl9yP')
        for i in brand:
            brand = i.text.strip()
            Brand_name.append(brand)
    except AttributeError:
        Brand_name.append('Not avialable')
    print(Brand_name)
    
    try:
        desc= box.find_all('a', class_ ='WKTcLC')
        desc
        for i in desc:
            desc=i.text
            Description.append(desc)
    except AttributeError:
        Description.append('Not avialable')
    print(Description)
    
    try:
        mrp= box.find_all('div', class_ ='yRaY8j')
        mrp
        for i in mrp:
            mrp=i.text
            MRP.append(mrp)
    except AttributeError:
        MRP.append('Not avialable')
    print(MRP)
    
    try:
        price= box.find_all('div', class_ ='Nx9bqj')
        price
        for i in price:
            price=i.text
            Price.append(price)
    except AttributeError:
        Price.append('Not avialable')
    print(Price)
    
    try:
        disc= box.find_all('div', class_ ='UkUFwK')
        disc
        for i in disc:
            disc=i.text[0:2]
            Discount.append(disc)
    except AttributeError:
        Discount.append('Not avialable')
    print(Discount)
    
    try:
        links = box.find_all('a', class_='rPDeLR')
        for i in links:
            href = i.get('href')
            full_link = 'https://www.flipkart.com' + href 
            Link.append(full_link)
    except AttributeError:
        Link.append('Not avialable')
    
    print(Link) 
```

```python
print(len(Brand_name))
print(len(Description))
print(len(MRP))
print(len(Price))
print(len(Discount))
print(len(Link))
```

```python
df = pd.DataFrame()
df['Brand_name']= Brand_name
df['Description']= Description
df['MRP']= MRP
df['Price']= Price
df['Discount']= Discount
df['Link'] = Link
```
